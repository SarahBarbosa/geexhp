## Generating random modern planets dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from geexhp.core import datavis, datagen, geostages
```

```python
dg = datagen.DataGen(url="http://127.0.0.1:3000/api.php")
```

```python
dg.generator(1_000, verbose=True, molweight=geostages.molweight_modern(), file="modern1k")
```

```python
df = pd.read_parquet("../data/modern1k.parquet")
```

```python
index = np.random.randint(0, len(df))
datavis.plot_spectrum(df, label=f"index={index}", index=index);
```
