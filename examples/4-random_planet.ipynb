## Generating random modern planets dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from geexhp.core import datavis, datagen, geostages
datavis.configure_matplotlib()
```

```python
start = 0
final = 8
dg = datagen.DataGen(url="http://127.0.0.1:3000/api.php", config = "../geexhp/config/default_habex.config")
```

```python
dg.generator(
    start=start,
    end=final,
    random_atm=False,
    verbose=True,
    file=f"modern_{start}-{final}",
    molweight=geostages.molweight_modern(),
)
```

```python
modern = pd.read_parquet("data/modern_0-8.parquet")
index = np.random.randint(0, len(modern))
datavis.plot_spectrum(modern, label=f"index={index}", index=index);
```

## Generating random after GOE planets dataset

```python
dg_goe = datagen.DataGen(url="http://127.0.0.1:3000/api.php",
                        config = "../geexhp/config/default_habex.config", stage="goe")
dg_goe.generator(
    start=start,
    end=final,
    random_atm=False,
    verbose=True,
    file=f"goe_{start}-{final}",
    molweight=geostages.molweight_after_goe(),
)
```
